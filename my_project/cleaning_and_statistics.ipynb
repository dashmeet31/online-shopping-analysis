```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("online_shopping_4000_instances.csv")

# Step 2: Preview the data
print("First 5 rows:")

print(df.head())

# Step 3: Dataset info
print("\nDataset Info:")
print(df.info())

# Step 4: Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Step 5: Statistical Summary
print("\nSummary Statistics:")
print(df.describe())

# Step 6: Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Step 7: Create new column for Total Amount after discount
df['Total Amount'] = df['Amount spent by per user'] - df['Discount']

# Step 8: Save cleaned data
df.to_csv("cleaned_online_shopping_data.csv", index=False)
print("\n Cleaned dataset saved successfully.")
```
